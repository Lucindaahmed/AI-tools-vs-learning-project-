This notebook contains all the code used to visualize the survey data collected for this project.

## Imports

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Load Data

```python
# Load the survey data
df = pd.read_csv('AI_tools_vs_effort (Responses) - Form Responses 1.csv')

df.head()
```

## Data Cleaning

```python
# Rename columns to shorter labels
df.columns = [
    'Timestamp', 'Status', 'Gender', 'Age', 'AI_Frequency', 'AI_Usage',
    'Time_Per_Day', 'Helpfulness', 'Learned', 'Easier_or_Faster',
    'Effort_Impact', 'Productivity', 'Faster_Tasks', 'Without_AI', 'Over_Reliance'
]

# Check for missing values
print("Missing values:")
print(df.isnull().sum())
print()
print(f"Total responses: {len(df)}")
```

## Descriptive Statistics

```python
# Mean, Median, Mode for numerical columns
for col in ['Helpfulness', 'Effort_Impact']:
    print(f"--- {col} ---")
    print(f"  Mean:   {df[col].mean():.2f}")
    print(f"  Median: {df[col].median():.2f}")
    print(f"  Mode:   {df[col].mode()[0]}")
    print()
```

## Visualizations
### Chart 1: Respondent Status

```python
counts = df['Status'].value_counts()

plt.bar(counts.index, counts.values, color=['#4F86C6', '#6DBF9E', '#F4A261'])
plt.title('Chart 1: Respondent Status Distribution')
plt.xlabel('Status')
plt.ylabel('Number of Respondents')

plt.show()
```

### Chart 2: AI Usage Frequency

```python
order = ['never', 'Rarely (1–2 times/month)', 'Sometimes (1–2 times/week)',
         'Often (3–5 times/week)', 'Very often (daily)']
labels = ['Never', 'Rarely', 'Sometimes', 'Often', 'Very Often']

counts = df['AI_Frequency'].value_counts().reindex(order, fill_value=0)

plt.bar(labels, counts.values, color=['#4F86C6', '#6DBF9E', '#F4A261', '#E76F51', '#A8DADC'])
plt.title('Chart 2: How Often Do Respondents Use AI Tools?')
plt.xlabel('Frequency of Use')
plt.ylabel('Number of Respondents')

plt.show()
```

### Chart 3: Helpfulness Rating (1-5)

```python
counts = df['Helpfulness'].value_counts().sort_index()

plt.bar([str(i) for i in counts.index], counts.values,
        color=['#4F86C6', '#6DBF9E', '#F4A261', '#E76F51', '#A8DADC'])
plt.title('Chart 3: How Much Do AI Tools Help With Understanding?\n(1 = Not Helpful, 5 = Extremely Helpful)')
plt.xlabel('Rating (1-5)')
plt.ylabel('Number of Respondents')

plt.show()

print(f"Mean: {df['Helpfulness'].mean():.2f} | Median: {df['Helpfulness'].median():.2f} | Mode: {df['Helpfulness'].mode()[0]}")
```

### Chart 4: Understanding vs. Effort (Pie Chart)

```python
counts = df['Easier_or_Faster'].value_counts()

plt.pie(counts.values,
        labels=counts.index,
        autopct='%1.1f%%',
        colors=['#4F86C6', '#F4A261', '#6DBF9E', '#E76F51'],
        startangle=140)
plt.title('Chart 4: Do AI Tools Improve Understanding or Just Reduce Effort?')

plt.show()
```

### Chart 5: Productivity Impact

```python
order = ['Much worse', 'Worse', 'Same', 'Better', 'Much better']
counts = df['Productivity'].value_counts().reindex(order, fill_value=0)

plt.bar(order, counts.values, color=['#E76F51', '#F4A261', '#A8DADC', '#6DBF9E', '#4F86C6'])
plt.title('Chart 5: How Do AI Tools Affect Productivity?')
plt.xlabel('Productivity Level')
plt.ylabel('Number of Respondents')

plt.show()
```

### Chart 6: Over-Reliance

```python
order = ['Strongly agree', 'Agree', 'Neutral', 'Disagree', 'Strongly disagree']
counts = df['Over_Reliance'].value_counts().reindex(order, fill_value=0)

plt.bar(order, counts.values, color=['#4F86C6', '#6DBF9E', '#F4A261', '#E76F51', '#A8DADC'])
plt.title('Chart 6: Do Students Rely Too Much on AI Tools?')
plt.xlabel('Level of Agreement')
plt.ylabel('Number of Respondents')

plt.show()
```

### Chart 7: Effort Impact Rating (1-5)

```python
counts = df['Effort_Impact'].value_counts().sort_index()

plt.bar([str(i) for i in counts.index], counts.values,
        color=['#4F86C6', '#6DBF9E', '#F4A261', '#E76F51', '#A8DADC'])
plt.title('Chart 7: How Much Do AI Tools Affect Study Effort?\n(1 = Still Work Hard, 5 = Rely Heavily on AI)')
plt.xlabel('Rating (1-5)')
plt.ylabel('Number of Respondents')

plt.show()

print(f"Mean: {df['Effort_Impact'].mean():.2f} | Median: {df['Effort_Impact'].median():.2f} | Mode: {df['Effort_Impact'].mode()[0]}")
```
